```python
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#new data set 
#filter out data 
df[['race', 'gender', 'age']]
df1=df[['patient_nbr','race', 'gender', 'age', 'time_in_hospital', 'number_diagnoses']]
display(df1)
```

```python
df1.dtypes
```

```python
#new data set: race and # of diagnoses 
df2= df1[(df1['race']== 'Hispanic')& (df1['number_diagnoses'] > 1) & (df1['age']== '[20-30)')]
display(df2)
```

```python
sns.pairplot(df2, hue = 'gender')
```

```python
#plot 
sns. barplot(x='number_diagnoses', y= 'time_in_hospital', data=df2)
#change tittle + labels 
plt.title("Number of Diagnoses vs Time in Hospital of Hispanic Diabetics (20-30 yrs old)")
plt.ylabel("Time in Hospital")
plt.xlabel ('Number of Diagnoses')
```

```python
sns.lmplot(x='number_diagnoses', y='time_in_hospital', hue='gender', data=df2)
#change tittle + labels 
plt.title("Number of Diagnoses vs Time in Hospital of Diabetic Hispanics in Relation to Gender (20-30 yrs old)")
plt.ylabel("Time in Hospital")
plt.xlabel ('Number of Diagnoses')
```

```python
#Calculare averge number of diagnoses for a Hispanic Male between the ages [20-30]
df3= df2[(df2['gender']== 'Male')]
display(df3)
df3.mean()
```

```python
#Calculare averge number of diagnoses for a Hispanic Female between the ages [20-30]
df4= df2[(df2['gender']== 'Female')]
display(df4)
df4.mean()
```
